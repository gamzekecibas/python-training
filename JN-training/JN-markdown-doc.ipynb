### Fundamental    
- New line: Double space end of the previous line.
- Alternative-1: < br > mark without spaces for new line.
- One indentation and dash is only dash.
- A text between double * or _ turns **bold.**
- Another alternative < b > and < /b > marks without spaces for bold.
- Just one * or _ mades *italic.*
- Another alternative < i > and < /i > marks without spaces for italic.

### Headings  
- There should be a space after last #.
Title : #Heading1   
Major heading: ##Heading2   
Subheadings: ###Heading3   
Title 4: ####Heading4    
Title 5: #####Heading5    
Title 6: ######Heading6   
### OR < h n > where n is order of the heading without spaces.

### Blockquotes
> This is a markdown documentation.
- Just ">" or < blockquote > (...) < /blockquote > marks without spaces

### Highlight
- To create highlight region, using < code > (...) < code > without spaces:  
<code>x=5  
y=pow(x,2)
    y=25<code>

### Mathematical Symbols  
- - Mathematical symbols are created by expression between dollar symbols.
- For all __[Mathematical Expressions](https://jupyterbook.org/content/math.html)__
- [Symbols List](https://www.w3schools.com/charsets/ref_utf_math.asp)  
- $\sqrt{2}$  
- &sum;  
- &forall;

### Horizontal Line
- Triple dash print a horizontal line as below:
---
- **or** < h r > mark without spaces.
<hr>

### Making List
---
- For ordered list:
    use just numbers 1., 2. etc.
    OR enter < o l > mark to initialize the list, enter each element between < li > and < /li > and finalize the list with < /ol >.
* For unordered lists, using dash, asterisk with space.

### Embedding Links
- Tag of the link is entered in square bracket and link is copied in paranthesis.
[Adding Link](https://stackoverflow.com/questions/25692293/inserting-a-link-to-a-webpage-in-an-ipython-notebook#:~:text=First%20open%20a%20markdown%20cell,to%20view%20in%20the%20Notebook.)  
- These tags can be made bold or italic.

### Images
- From same file:  
      ! [< caption >] (image_filename.png)
- From another file:   
      ! [< caption >] (folder/image_filename.png)    
    without spaces.
- OR just drop the image in markdown cell.
